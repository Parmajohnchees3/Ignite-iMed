```python
import pandas as pd 
import numpy as np

from sklearn import preprocessing
from sklearn.neural_network import MLPClassifier
```

```python
df_train = pd.read_csv("data/Training.csv")
df_train.drop(df_train.columns[len(df_train.columns)-1], axis=1, inplace=True)
df_test = pd.read_csv("data/Testing.csv")
```

```python
df_train
```

```python
y_train = df_train["prognosis"]
X_train = df_train.drop("prognosis", axis=1)

y_test = df_test["prognosis"]
X_test = df_test.drop("prognosis", axis=1)
```

```python
le = preprocessing.LabelEncoder()
le.fit(y_train)

y_train = le.transform(y_train)
y_test = le.transform(y_test)
```

```python
y_test
```

```python
neural_net = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(5, 2), random_state=1)

neural_net = neural_net.fit(X_train, y_train)
```

```python
results = neural_net.predict(X_test)
```

```python
neural_net.score(X_test, y_test)
```
